```python
import matplotlib.pyplot as plt
import math as ma
import numpy as np
import matplotlib.pyplot as plt
def func(x):
    return ma.log(x+ma.sin(x**2/100))
f=np.vectorize(func)
x=np.arange(1,100,0.1)
y=f(x)
fig = plt.figure()
plt.plot(x, y)
plt.show()
x0=np.arange(1,100,3)
y0=f(x0)
z=np.vstack((x0, y0))
z=np.transpose(z)
np.savetxt("asdasd.txt", z)
plt.plot(x, y, x0,y0,".")
plt.show()
```

```python
import scipy.interpolate as sc
z= np.loadtxt('asdasd.txt')
x=z[:,0]
y=z[:,1]
yint=sc.interp1d(x,y,kind='cubic')
yintlin=sc.interp1d(x,y,kind='linear')
xint=np.linspace(1,97,20)
plt.plot(xint,yint(xint),'--',label='interpolation linear')
plt.plot(xint,yint(xint),'-',label='interpolation cubic')
plt.plot(x,y,'.',label='data')
plt.legend()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f1(a,b,c,x):
    return (a*x**2+b*x+c)
```

```python
def para(a,b,c,start,end):
    x=np.linspace(start, end, 10**6)
    f=list()
    for w in x:
        f.append(f1(a,b,c,w))
    plt.plot(x, f, '-')
```

```python
para(1,4,4,-4,0)
```

```python
para(1,4,3,-4,0)
```

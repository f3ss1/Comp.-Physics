```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f1(x):
    return np.cos(x)
def f2(x):
    return -np.sin(x)
```

```python
def der(start, end):
    x = np.linspace(start, end, 10**6)
    fp = list()
    fr = list()
    for w in x:
        fp.append((f1(w+10**(-6))-f1(w))/(10**(-6)))
        fr.append(f2(w))
    plt.plot(x, fp, '-', x, fr, '-')
```

```python
der(-np.pi,np.pi)
```

```python
der(-4*np.pi,4*np.pi)
```
